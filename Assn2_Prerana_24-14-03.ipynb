```python
#d)
mean = np.mean(data, axis=0)
std_dev = np.std(data, axis=0)
print("\nMean of each column:\n", mean)
print("\nStandard deviation of each column:\n", std_dev)
```

```python
#e)
normalized = (data - mean) / std_dev
print("\nStandardized data (normalized):\n", normalized)
```

```python
#3rd question
#a)
import numpy as np
import numpy.linalg as la
def vandermonde(N):
 vec = np.arange(N) + 1
 vander = vec[:, np.newaxis] ** vec[np.newaxis, :]
 return vander
N = 12
vander = vandermonde(N)
print("Vandermonde matrix:")
print(vander)
```

```python
#b)
x = np.ones(N)
b = vander @ x
print("Vector b:")
print(b)
```

```python
#c)
x_naive = la.inv(vander) @ b
print("Naive solution:\n", x_naive)
```

```python
#d)
x_solve = la.solve(vander, b)
print("Solution using np.linalg.solve:", x_solve)
```
